```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
import scipy.stats as stats
from sklearn.naive_bayes import GaussianNB # The distribution looks like a gaussian... ?
from sklearn.naive_bayes import MultinomialNB
import sklearn.preprocessing as pre
```

```python
df = pd.read_csv("500_Person_Gender_Height_Weight_Index.csv")
df['BMI'] = df['Weight'] / ((df['Height']/100)**2)
```

```python
mylist = [1.0 if i == "Male" else 0.0 for i in df['Gender']]
df['Sex'] = mylist
```

## Single variable Gaussian NB using BMI to predict gender.

```python
X = df['BMI'].to_numpy()
y = df['Sex']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
GaussNB = GaussianNB()
GaussNB.fit(X_train.reshape(-1,1), y_train)
y_preds = GaussNB.predict(X_test.reshape(-1,1))
```

```python
y_preds
```

```python
y_test.to_numpy()
```

```python
GaussNB.predict([[38.0]]) # This is the approximate cutoff for classifying Male: 38 and higher.
```

```python
accuracy_score(y_test.to_numpy(), y_preds)
```

```python
# What if we had randomly guessed female?
255/500
```

```python
mNB = MultinomialNB()
mNB.fit(X_train.reshape(-1,1), y_train)
y_preds2 = mNB.predict(X_test.reshape(-1,1))
```

```python
accuracy_score(y_test.to_numpy(), y_preds2)
```

```python
# The accuracy score of 0.55 is not great, considering a 0.51 guessing accuracy score. But it's close to the 
# theoretical maximum for this dataset, given that the distributions are not well-separated.
```

## Multivariate Naive Bayes?

```python
# myscaler = pre.StandardScaler(copy=False). # on second thought, I don't think this is needed.
X2 = df[['Height','Weight']].to_numpy() #myscaler.fit_transform(df[['Height','Weight']].to_numpy())
y2 = df['Sex']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.4, random_state=42)
```

```python
mNB2 = MultinomialNB()
mNB2.fit(X2_train, y2_train)
y_pred2 = mNB2.predict(X2_test)
```

```python
accuracy_score(y2_test.to_numpy(), y_pred2)
```
