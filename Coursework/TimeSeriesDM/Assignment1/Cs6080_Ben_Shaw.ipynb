**Question 5**: Implement two functions called get_MovingAverage and get_MovingMedian that takes 2 arguments:
- data: the original data time series
- k : that helps you find how many elements the MA and MM should consider

The functions should return the detrended data 

**Remember:** the number of points considered in every window of the moving average is 2K+1

```python
def get_MovingAverage(data, K):
    #detrended = []
    trended = []
    L = len(data)
  #detrended = []
  # INSERT CODE HERE.   To get the average of a list (for example [1,2,3]) use np.mean([1,2,3])
    # we note that mt = sum(1/(2k+1) x_{t+i}, i=-k..k). But what if "i" is too small or large?
    for i in range(L):
        if (i >= K) and (i <= (L-K)-1): #This is the typical case.
            sublist = data[i-K:i+1+K]
            trended.extend([np.mean(sublist)])
        elif i < K: #What if the data point is closer than K to the beginning? We could get nan values, or...
            sublist = data[0:2*i+1]
            trended.extend([np.mean(sublist)])
        else: #Same issues here. Why get nan values when you can take a crack at what the trend might look like?
            d = L-i
            sublist = data[L-2*d+1:L]
            trended.extend([np.mean(sublist)])
    
    detrended = [data[i:i+1].values[0][0]-trended[i] for i in range(L)] #detrend.
    return detrended
  #return detrended


# When done, uncomment the code below, you should get the same results of the TSA library
detrended = get_MovingAverage(data, 10)
plt.plot(detrended)
plt.plot(detrend_data) # I added this part to show that the main difference is my not giving up datapoints
plt.show()
```

```python
def get_MovingMedian(data, K):
  #detrended = []
  # INSERT CODE HERE.  To get the median of a list (for example [1,2,3])  use np.median([1,2,3])
    #detrended = []
    trended = []
    L = len(data)
    # we note that mt = sum(1/(2k+1) x_{t+i}, i=-k..k). But what if "i" is too small or large?
    for i in range(L):
        if (i >= K) and (i <= (L-K)-1):
            sublist = data[i-K:i+1+K]
            trended.extend([np.median(sublist)])
        elif i < K:
            sublist = data[0:2*i+1]
            trended.extend([np.median(sublist)])
        else:
            d = L-i
            sublist = data[L-2*d+1:L]
            trended.extend([np.median(sublist)])
    
    detrended = [data[i:i+1].values[0][0]-trended[i] for i in range(L)] #detrend
    return detrended
  #return detrended

# When done, uncomment the code below
detrended = get_MovingMedian(data, 10)
plt.plot(detrended)
plt.show()
```

**Question 6**: Implement the exponential smoothing we have seen in class with alpha=0.8.

```python
def get_ExponentialSmoothning(data, alpha=.8):
    #detrended = []
    trended = []
    L = len(data)
    for i in range(L):
        if i == 0:
            trended.extend([data.values[i][0]])
        else:
            term = alpha*data.values[i][0] + (1-alpha)*trended[i-1]
            trended.extend([term])
    
    detrended = [data[i:i+1].values[0][0]-trended[i] for i in range(L)] #detrend.
    return detrended
  #return detrended
  
detrended = get_ExponentialSmoothning(data)
plt.plot(detrended)
plt.show()
```
