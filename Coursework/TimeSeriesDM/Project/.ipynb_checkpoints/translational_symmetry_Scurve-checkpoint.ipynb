```python
def TSrep(v,X,chunks=10):
    #first get the matrix.
    A1 = np.array([[component for component in v]])
    A = scipy.linalg.null_space(A1)
    At = A.transpose()
    P = A@At #Note that the columns of A are orthonormal (approximately)
    Xt = X.transpose()
    Pb = P@Xt
    d2P = At@Xt #This will be the projection used later on.
    d2Pt = d2P.transpose()
    N = Xt-Pb
    Nt = N.transpose()
    N2 = Nt@N #the diagonals should be the dist squared from the point to the plane.
    dists = [np.sqrt(N2[i][i]) for i in range(N2.shape[0])]
    maxDist = max(dists)
    vn = v/np.linalg.norm(v)/chunks
    lengthOfT = 2*maxDist           #Now we're going to chunk up the data based on the distance.
    pieceLength = lengthOfT/chunks
    Lendpoints = [i*pieceLength for i in range(chunks)]
    #print(Lendpoints)
    
    def getSign(vn,x):  #What side of vn is the datapoint on?
        w = x-vn #This is what connects x and vn
        proj = (w@vn)*chunks #Now we check the scaler projection of w onto vn.
        if proj > pieceLength:
            return 1
        elif proj < -pieceLength:
            return -1
        else:
            return 0 #"no man's land:" in between.
    
    
    mydict = {}
    for pt in Lendpoints:
        mydict[pt] = []
        
    #print(mydict)
    #print(vn)
    def bindist(d,x,dic):
        #ID = iter(dic)
        diclist = [key for key in dic]
        pkey = diclist[0]
        lkey = diclist[-1]
        for i in range(1,len(diclist)):
            nkey = diclist[i]
            if d>=pkey and d<nkey:
                dic[pkey].extend([x])
                #dic[pkey].extend([d])
                return
            pkey = diclist[i]
        dic[lkey].extend([x])
        #dic[lkey].extend([d])
        
    
    for i in range(X.shape[0]): #Now we bin the data. each bin is a "timestamp"
        x = X[i]
        p = d2Pt[i]
        dist = dists[i]
        sign = getSign(vn,x)
        #print(sign)
        if sign > 0:
            binval = maxDist*dist
            bindist(binval,p,mydict)
            #print(sign)
            
        elif sign < 0:
            binval = maxDist - dist
            #print(binval)
            bindist(binval,p,mydict)
        else:
            binval = dist
            #print(binval)
            bindist(binval, p, mydict)
            
    bigX = [[np.round(key,4),np.array(mydict[key])] for key in mydict]
    
    #for key in mydict:
     #   print(mydict[key])
            
    return bigX
```

```python
def getParams(X,f,DM,d):
    dM = eval(DM + "(n_components=" + str(d) + ")")
    #print(dM)
    trans = dM.fit_transform(X)
    transT = trans.transpose()
    tvals = transT[0]
    # first we need to standardize.
    xmin = min(tvals)
    xmax = max(tvals)
    mid = (xmin + xmax)/2
    #first slide things over.
    tvals = [t-mid for t in tvals]
    #now make the range be from -1 to 1.
    stFac = 1/(xmax-mid)
    tvals = [t*stFac for t in tvals]
    Xt = X.transpose()
    funcList = []
    #plt.scatter(tvals,Xt[0])
    #plt.scatter(tvals,Xt[1])
    for x in Xt:
        popt1, pcov1 = curve_fit(f, tvals, x)
        #paramsList.extend([popt1])
        X1 = lambda x: f(x, *popt1)
        funcList.extend([X1])
        
    return funcList
```

```python
myAve = getParams(bigChunk[0][1],myfunc,"SpectralEmbedding",2)
```

```python
myts = [getParams(thing[1],myfunc,"SpectralEmbedding",2) for thing in mychunks if len(thing[1])>0]
```

```python
def IP(r1,r2):
    Dfunc = lambda x: r1[0](x)*r2[0](x) + r1[1](x)*r2[1](x)
    term = integrate.quad(Dfunc,0,2*np.pi)
    return term[0]

def angle(r1,r2):
    term = IP(r1,r2)/np.sqrt(IP(r1,r1)*IP(r2,r2))
    theta = np.arccos(term)
    return theta
```

```python
angles = [angle(myAve,r) for r in myts]
sims = [1-ang/np.pi for ang in angles]
ave = sum(sims)/len(sims)
print(ave)
```

```python
#This is the function which will tie all of this in to estimate the similarity along an axis.
def getAveSim(v,X,f,dm="SpectralEmbedding",D=2,num_chunks=10):
    bigChunk = TSrep(myV,X2,chunks=1)
    mychunks = TSrep(v,X,chunks=num_chunks)
    myAve = getParams(bigChunk[0][1],f,dm,D)
    myts = [getParams(thing[1],f,dm,D) for thing in mychunks if len(thing[1])>0]
    angles = [angle(myAve,r) for r in myts]
    sims = [1-angle/np.pi for angle in angles]
    ave = sum(sims)/len(sims)
    return ave
```

```python
aS2 = getAveSim(myBadV1,X2,myfunc)
aS3 = getAveSim(myBadV2,X2,myfunc)
print(aS2)
print(aS3)
```
