## Can PCA be used to help us know what symmetries to test for?

```python
# Prediction: the principal components are the x, y, and z axes.
pca = PCA(n_components=3)
pca.fit(X2)
```

```python
pca.singular_values_
```

```python
# It is thought that if the dataset exhibits translational symmetry, it is likely in the direction
# of one of the principal components. This is probably not true in general, but it may serve as a 
# useful idea for testing for symmetries in practice.
```

```python
pca.components_
```

```python
# Here's the principal component which most closely corresponds to the known direction of translational symmetry.
myV = pca.components_[2]
myV
```

```python
# Here will be two "bad" directions of translational symmetry, for later comparison.
myBadV1 = pca.components_[1]
myBadV2 = pca.components_[0]
```

```python
# This function will essentially break up the dataset into "level sets" with respect to a given direction.
# It's possible that this function can be simplified and otherwise improved upon.

def TSrep(v,X,chunks=10):
    #first get the matrix.
    A1 = np.array([[component for component in v]])
    A = scipy.linalg.null_space(A1)
    At = A.transpose()
    P = A@At #Note that the columns of A are orthonormal (approximately)
    Xt = X.transpose()
    Pb = P@Xt
    d2P = At@Xt #This will be the projection used later on.
    d2Pt = d2P.transpose()
    N = Xt-Pb
    Nt = N.transpose()
    N2 = Nt@N #the diagonals should be the dist squared from the point to the plane.
    dists = [np.sqrt(N2[i][i]) for i in range(N2.shape[0])]
    maxDist = max(dists)
    vn = v/np.linalg.norm(v)/chunks
    lengthOfT = maxDist #2*maxDist           #Now we're going to chunk up the data based on the distance.
    pieceLength = lengthOfT/chunks
    Lendpoints = [i*pieceLength for i in range(chunks)]
    #print(Lendpoints)
    
    def getSign(vn,x):  #What side of vn is the datapoint on?
        w = x-vn #This is what connects x and vn
        proj = (w@vn)*chunks #Now we check the scaler projection of w onto vn.
        if proj > pieceLength:
            return 1
        elif proj < -pieceLength:
            return -1
        else:
            return 0 #"no man's land:" in between.
    
    
    mydict = {}
    for pt in Lendpoints:
        mydict[pt] = []
        
    #print(mydict)
    #print(vn)
    def bindist(d,x,dic):
        #ID = iter(dic)
        diclist = [key for key in dic]
        pkey = diclist[0]
        lkey = diclist[-1]
        for i in range(1,len(diclist)):
            nkey = diclist[i]
            if d>=pkey and d<nkey:
                dic[pkey].extend([x])
                #dic[pkey].extend([d])
                return
            pkey = diclist[i]
        dic[lkey].extend([x])
        #dic[lkey].extend([d])
        
    
    for i in range(X.shape[0]): #Now we bin the data. each bin is a "timestamp"
        x = X[i]
        p = d2Pt[i]
        dist = dists[i]
        sign = getSign(vn,x)
        #print(sign)
        if sign > 0:
            binval = maxDist*dist
            bindist(binval,p,mydict)
            #print(sign)
            
        elif sign < 0:
            binval = maxDist - dist
            #print(binval)
            bindist(binval,p,mydict)
        else:
            binval = dist
            #print(binval)
            bindist(binval, p, mydict)
            
    bigX = [[np.round(key,4),np.array(mydict[key])] for key in mydict]
    
    #for key in mydict:
     #   print(mydict[key])
            
    return bigX
```

```python
# Here we'll break up the s-curve into level sets with respect to the known good direction.
mychunks = TSrep(myV,X2)
# len(mychunks)
```

```python
# How many data points are in each "level set?"
[mychunks[i][1].shape for i in range(len(mychunks))]
```

```python
# Define a function to plot the "level sets."
def plotChunk(chunk):
    Xt = chunk.transpose()
    x = Xt[0]
    y = Xt[1]
    plot = plt.scatter(x,y)
    return plot
```

```python
plotChunk(mychunks[0][1])
```

```python
plotChunk(mychunks[1][1])
```

```python
# How can we quantify the similarity between two "level sets?" Here I aggregate the 1-dimensional
# wasserstein distance (known as the Earth mover's distance, I believe) using the square root of the 
# sum of squares. There is probably a much better idea for comparing level sets.

def dist(M1,M2):
    xdist = scipy.stats.wasserstein_distance(M1.transpose()[0],M2.transpose()[0])
    ydist = scipy.stats.wasserstein_distance(M1.transpose()[1],M2.transpose()[1])
    dist = np.sqrt(xdist**2 + ydist**2)
    return dist
```

```python
# Now let's create a heatmap which shows us the distance between each of the level sets.
mymat = []
for i in range(10):
    thing = []
    for j in range(10):
        thing += [dist(mychunks[i][1], mychunks[j][1])]
        
    mymat += [thing]
    thing = []

mymat = np.array(mymat)
```

```python
sns.heatmap(mymat, linewidth=0.5)
plt.show()
# mymat
```
